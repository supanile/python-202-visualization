```python
import matplotlib.pyplot as plt

# ข้อมูลตัวอย่าง
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# ปรับแต่งสีเส้นและรูปแบบเส้น
plt.plot(x, y, color='red', linestyle='--', marker='o', markerfacecolor='blue', markersize=10)

# เพิ่ม title และ label
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# แสดงกราฟ
plt.show()
```
